```python
# imports
import openai
import json
from IPython.display import Markdown, display, update_display
```

```python
# constants

MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'

MODEL = MODEL_GPT
```

```python
system_prompt = """
You are technical assistance that can answer any quistions that might come. The quiestions should be answered with a short explanation and then a short example. 
The person you are assisting is just above high school level. 
"""
```

```python
# set up environment
def ai_assistance(message : str) -> None:
    
    stream = openai.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": message}
          ],
        stream=True
    )    
    print(f"Streaming the brochure")
    response = ""
    display_handle = display(Markdown(""), display_id=True)
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        update_display(Markdown(response), display_id=display_handle.display_id)
```

```python
# Here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
# Get gpt-4o-mini to answer, with streaming

ai_assistance(question)
```

```python
# Get Llama 3.2 to answer

# Run 'ollama serve' if it is not already running

from openai import OpenAI
openai = OpenAI()
OLLAMA_BASE_URL = "http://localhost:11434/v1"

ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
def ai_assistance_ollama(message : str) -> None:
    messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": message}
          ]
    response = ollama.chat.completions.create(model="llama3.2:1b", messages=messages)

    print(response.choices[0].message.content)
```

```python
ai_assistance_ollama(question)
```
